```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
df = sns.load_dataset('iris')

# Preview the first few rows
print("--- Dataset Head ---")
print(df.head())

# Check data types and structure
print("\n--- Info ---")
df.info()
```

```python
# Check for nulls
missing_values = df.isnull().sum()
print("\n--- Missing Values Per Column ---")
print(missing_values)

# Logic: If missing values existed, we could fill them with the mean 
# (only for numeric columns) or drop them.
# df.fillna(df.mean(), inplace=True)
```

```python
print("\n--- Summary Statistics ---")
print(df.describe())
```

```python
sns.pairplot(df, hue='species', palette='viridis', diag_kind='kde')
plt.show()
```

```python
plt.figure(figsize=(12, 6))

# Melt the dataframe to make it "long-form" for easier plotting
df_melted = df.melt(id_vars="species", var_name="Feature", value_name="Value")

sns.boxplot(data=df_melted, x="Feature", y="Value", hue="species")
plt.title("Feature Distribution by Species")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.violinplot(x='species', y='sepal_length', data=df, inner="quartile")
plt.title("Sepal Length Density by Species")
plt.show()
```
